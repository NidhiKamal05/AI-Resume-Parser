```python
import fitz # PyMuPDF
import spacy
import gradio as gr

# print("Environment Ready!")
```

```python
def extract_text_from_pdf(pdf_file):
  """
  PDF text reader
  """
  
  try:
    doc = fitz.open(pdf_file.name)

    text = ""
    for page in doc:
      text += page.get_text()

    cleaned_text = " ".join(text.split())

    if not cleaned_text.strip():
      return "Error: No text found, either it's not in pdf form or it's scanned image"

    return cleaned_text

  except Exception as e:
    return f"Error occurred: {str(e)}"
```

```python
interface = gr.Interface(
    fn = extract_text_from_pdf,
    inputs = gr.File(label = "Upload Resume (PDF only)", file_types = [".pdf"]),
    outputs = gr.Textbox(label = "Extracted Raw Text", lines=  15),
    title = "AI Resume Parser",
    description = "Upload resume in PDF format to be read by AI."
)
```

```python
interface.launch(debug=True)
```
